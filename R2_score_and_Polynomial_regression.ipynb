```python
y = df.iloc[:,2].values
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x,y)
```

```python
y_pred = regressor.predict(x)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
tem = df['tem'].tolist()
pre = df['pres'].tolist()

plt.plot(tem,pre,color = 'blue')
plt.scatter(tem,y_pred,color = 'red')

plt.xlabel('temperature')
plt.ylabel('pressure')
plt.xticks(pre)
plt.title('simple linear regression')

plt.legend()
plt.show()
```
